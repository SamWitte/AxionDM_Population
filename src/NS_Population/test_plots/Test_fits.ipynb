```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
file_test = np.loadtxt("../Test_Run.dat")
```

```python
Plist = np.unique(file_test[:,0])
Blist = np.unique(file_test[:,1])
P_s_list = np.unique(file_test[:,2])
B_s_list = np.unique(file_test[:,3])
#
fig, axs = plt.subplots(5,5, figsize=(20, 20))
temp = []
for i in range(len(P_s_list)):
    for j in range(len(B_s_list)):
        Ps_idx=i
        Bs_idx=j

        cond1 = file_test[:,2] == P_s_list[Ps_idx]
        cond2 = file_test[:,3] == B_s_list[Bs_idx]
        zz = file_test[np.all(np.column_stack((cond1,cond2)),axis=1), -1].reshape(len(Plist),len(Blist)) 
        # print(i,j, np.max(zz))
        axs[i,j].contourf(Plist, Blist, np.log10(zz), levels=np.array([-10, -5, -4, -3, -2, -1]))
        axs[i,j].text(-.4, 13.2, "sP {:.2f}\n sB {:.2f}".format(P_s_list[Ps_idx], B_s_list[Bs_idx]))
        temp.append([i,j,np.max(zz)])
temp = np.asarray(temp)
print(temp[np.argmax(temp[:,-1])])
```
